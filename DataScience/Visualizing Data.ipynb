```python
from matplotlib import pyplot as plt
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
#add a title
plt.title('Nominal GDP')
#labels on x and y axis
plt.ylabel("Millions in Rupees")
plt.xlabel("Years")
# create a line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color = 'red', marker = 'o', linestyle = 'solid')
plt.show()
```

```python
movies = ['Annie Hall', 'Ben-Hur', 'Casablanca', 'Gandhi', 'West Side Story']
num_oscars = [5, 11, 3, 8, 10]
```

```python
# plot bars with left x-coordinates [move names], heights [num_oscars]
plt.bar(movies, num_oscars)
plt.title("Oscar Winning Movies")
plt.ylabel("Number of Oscars")
plt.xlabel("Movie Names")
```
